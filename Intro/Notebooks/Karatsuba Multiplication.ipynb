# Karatsuba Multiplication

## Abstract
Multiplying large numbers can require a lot of compute power. Is there a better way? The answer is Recursion

What if we can split the number in half? Then we have a forumla; 
        
        int(str(a)+str(b))*int(str(c)+str(d))= ac10^m + ((a+b)(c+d)-ac-bd)10^m +bd
        
        for Example: the numbers 1234 and 5678 would get split into a=12, b=34, c=56, d=78
        and the result would be 7006652

After we can evaluate how many terms each half has.
- If it had 1 term, or a value < 10, then we can finish the recussive call and terminate the program (since its now single integer multiplication)
- Else we can make a recursive call to our function, resplitting the numbers and checking

```python
import time
import numpy as np
```

```python
k=1250
rando = np.random.rand(k,2)
x=''
y=''

for i in range(k):
    if i <= k//2:
        x+=str(int(rando[i][0]*10000000000000))+str(int(rando[i][1]*10000000000000))
    else:
        y+=str(int(rando[i][0]*10000000000000))+str(int(rando[i][1]*10000000000000))
        
x,y = int(x),int(y)
print('x is: '+str(x))
print('y is: '+str(y))
```

```python
def karat(x,y):
    if x < 10 or y < 10: # checck if either of the 
        return x*y
    else:
        m = max(len(str(x)),len(str(y)))
        m2 = m // 2

        a = x // 10**(m2)
        b = x % 10**(m2)
        c = y // 10**(m2)
        d = y % 10**(m2)

        z0 = karat(b,d)
        z1 = karat((a+b),(c+d))
        z2 = karat(a,c)

        return (z2 * 10**(2*m2)) + ((z1 - z2 - z0) * 10**(m2)) + (z0)
```

```python
start = time.perf_counter()
karat(x,y)
end = time.perf_counter()
print('Karatsu Runtime: '+str(end-start)+' Seconds')
```
