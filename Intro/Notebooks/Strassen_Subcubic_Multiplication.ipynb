# Strassen's Subcubic Matrix Multiplication Algorithm 


### The Problem

    Matrix Multiplication X*Y = Z, where X, Y are n x n matricies
    Input size: O(n^2)
    
    Z_ij = (ith row of X)(jth column of Y)
    Z = X*Y = SUM(k =1 => n){(X_ik)(Y_kj)}
    
    ex n=2: (a b) (e f) = (ae+bg af+bh)
            (c d) (g h) = (ce+dg cf+dh)
            
            
    Runtime: O(n^3)
        Proof: z_ij = (ith row of X)(jth column of Y), which 
        will run in O(n) time and there are O(n^2) terms
    
#### Fitting Divide and Conquer Paradigm    
    1) Divide into smaller subproblems
    
        Matrix) Divide X into quadrants 
                (a b)
                (c d)
                
                Divide Y into quadrants
                (e f)
                (g h)
    
    2) Conquer subproblems recursively
    
        Matrix) X*Y = (ae+bg af+bh)
                      (ce+dg cf+dh)
    
    3) Combine solutions of subproblems into the orignial
        

##### Recursive Algorithm #1
    Step 1: Recusrively compute 8 necessary products
    Step 2: Do the necessary addtions O(n^2) time
        => Runtime is O(n^3)

## Strassen's Algorithm (1969)

    Step 1: Recursivly compute 7 products
    
        Matrix) Divide X into quadrants 
                (a b)
                (c d)
                
                Divide Y into quadrants
                (e f)
                (g h)
                
                p1=a(f-h)
                p2=(a+b)h
                p3=(c+d)e
                p4=d(g-e)
                p5=(a+d)(e+h)
                p6=(b-d)(g+h)
                p7=(a-c)(e+f)
    Step 2: Conquer subproblems recursively
    
        Matrix) X*Y = (ae+bg af+bh)= (p5+p4-p2+p6       p1+p2)
                      (ce+dg cf+dh)= (p3+p4       p1+p5-p3-p7)
                      
            Proof: check algebra, not intuitive
            
    Runtime: O(n^log7) ~ O(n^2.8074)
